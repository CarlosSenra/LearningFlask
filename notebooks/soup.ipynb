```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from typing import List
import logging
```

```python
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

logger = logging.getLogger('BookScraping')
```

## Creating scraping class

```python
class BookScraping():
    """Execute web screpping from base url 'https://books.toscrape.com/'."""
    def __init__(self):
        self.base_url = "https://books.toscrape.com/"
        
        try:
            logger.info('Making get request.')
            self.home_page_response = requests.get(self.base_url)
        except Exception as e:
            logger.error(f'Error to get request from {self.base_url}: {e}')
            
    def get_ancoras(self):
        """Get all anchors from home page ulr base"""
        try:
            logger.info(f'Searching all anchors from base url {self.base_url}.')
            soup = BeautifulSoup(self.home_page_response.text, 'html.parser')
            self.home_anchors = soup('a')
            logger.info(f'Finded {len(self.home_anchors)} from base url.')
        except Exception as e:
            logger.error(f'Error to Searching all anchors: {e}')
        
    def create_list_catageroies_books(self):
        """Select all href from home page books and put all in a list"""
        self.list_categories_books = []
        for anchor in self.home_anchors:
            ref = anchor.get('href')
            if 'catalogue/category/' in ref:
                self.list_categories_books.append(ref)
        self.list_categories_books
    
    def request_book_category_page(self, book_category_href: str) -> List:
        """Make a request in category books page an get all books 

        Args:
            book_category_href (str): category href from base url.

        Returns:
            List: A list of all books in a catogry
        """
        _book_list = []
        _response = requests.get(self.base_url+book_category_href)
        _soup = BeautifulSoup(_response.text)
        articles = _soup.find_all("article", class_="product_pod")
        for article in articles:
            _book_list.append(article)
        return _book_list
    
            
    def print_something(self):
        self.get_ancoras()
        self.create_list_catageroies_books()
        book_list = self.request_book_category_page(book_category_href=self.list_categories_books[1])
        print(book_list[0])
        #print(self.list_categories)
    
scrap = BookScraping()
scrap.print_something()
```
